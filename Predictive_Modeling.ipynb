**Step1: Importing CSV DATA, Example: import_data.csv**

```python
# @title Default title text
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset (assuming it's in CSV format)
data = pd.read_csv('Import_data.csv')
data.head()
```

**Step 3: Visualizing Trade Trends**

```python
# Filter data for specific countries
china_data = data[data['Country'] == 'China']
pakistan_data = data[data['Country'] == 'Pakistan']

# Plot trade trends
plt.figure(figsize=(12, 6))
plt.plot(china_data['Year'], china_data['Import value of each commodity'], label='China', marker='o')
plt.plot(pakistan_data['Year'], pakistan_data['Import value of each commodity'], label='Pakistan', marker='o')
plt.xlabel('Year')
plt.ylabel('Trade Volume (in millions)')
plt.title('Indias Trade Volume with China and Pakistan')
plt.legend()
plt.show()
```

**Step 2: Data Cleaning and Exploration**

```python
# Check for missing values
print(data.isnull().sum())

# Drop any rows with missing values, if necessary
data = data.dropna()

# Quick overview of trade volume by country
trade_by_country = data.groupby('Country')['Import value of each commodity'].sum().sort_values(ascending=False)
print(trade_by_country)
```
